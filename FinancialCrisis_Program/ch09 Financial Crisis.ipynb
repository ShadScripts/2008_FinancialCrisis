# Financial Crisis Data and Drawing on a Subplot

The dataset contains closing S&P 500 Index price since 2007 obtained from Yahoo! Finance.  


```python
import matplotlib.pyplot as plt
import numpy as np
from numpy.random import randn
import pandas as pd
from datetime import datetime
plt.rc('figure', figsize=(10, 6))
```

```python
data = pd.read_csv('spx.csv', index_col=0, parse_dates=True) 
```

```python
#spx is a Series object
spx = data['SPX']
```

### S&P visual with annotations:

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
spx.plot()

#notable events
crisis_data = [
    (datetime(2007, 10, 11), 'Peak of bull market'),
    (datetime(2008, 3, 12), 'Bear Sterns fails'),
    (datetime(2008, 9, 15), 'Lehman Bankruptsy')]

#iterating through crisis data and annotating with arrows
for (date, label) in crisis_data:
    ax.annotate(label, 
                xy=(date, spx.asof(date)+ 65 ),
                xytext=(date, spx.asof(date)+ 200 ),
                arrowprops=dict(headwidth=4, width=1, headlength=10),
                horizontalalignment='left', verticalalignment='top')

    
# Zoom in on 2007-2010
ax.set_xlim(['1/1/2007', '1/1/2011'])
ax.set_ylim([600, 1800])

ax.set_title('Dates in the 2008-2009 Financial Crisis')
```
